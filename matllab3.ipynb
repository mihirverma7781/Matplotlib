```python
from matplotlib import pyplot as plt

slices = [30,20,50,10,5]
labels = ['java','python','javascript','flutter','c++']
colors = ['red','green','blue','yellow','cyan']
explode = [0,0,0.1,0,0]

plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'white'},shadow='true',autopct='%1.1f%%')



plt.title('Pie chart')
plt.tight_layout()
plt.show()
```
